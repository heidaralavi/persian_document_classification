```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('sample.csv')
```

```python
data['شرح'][0]
```

```python
sample_persian= data['شرح'][0]
```

```python
type(sample_persian)
```

```python
import re
#cleaning text
sentences = sample_persian
corpus = []

for sent in sentences:
    review = re.sub("[^a-zA-Z]", " ", sent)
    review = re.sub("\b[a-zA-Z]\b", " ", review)
    review = review.lower()
    review = review.split()
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
import re
#cleaning text
sentences = sample_persian
corpus = []

for sent in sentences:
    if re.search('[\u0600-\u06FF]',sent):
        corpus.append(sent)
```
